### Pie plots | Scatter Plots |  Bubble Plots..

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
df = pd.read_csv('canada.csv',index_col=0)
df.head()
```

```python
print(df.shape)
```

```python
df =df.drop(columns=['Type','Coverage','REG','AREA','DEV'],axis=1)
df.head()
```

```python
df.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)
```

```python
df.head()
```

```python
years = list(map(str,range(1980,2014)))
df.columns = list(map(str,df.columns))
df.set_index('Country',inplace=True)
df['Total'] = df.sum(axis=1)
```

```python
df.head(2)
```

## Pie Charts

```python
mpl.style.use('ggplot')
```

```python
df_continent = df.groupby('Continent',axis=0).sum()
df_continent.head(2)
```

```python
df_continent['Total'].plot(kind='pie',figsize=(10,6),autopct='%1.1f%%',startangle=10,shadow=True)
plt.title('No of Imigrants to canada by contient from [1980 - 2013]')
plt.axis('equal')
plt.show()
```

### Lets make it more clear

```python
colors_list = ['gold', 'yellow', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list =[0.2,0.1,0.1,0.2,0.2,0.2]

df_continent['Total'].plot(kind='pie',figsize=(15,7),
                           colors =colors_list,autopct='%1.1f%%',
                           startangle=50,shadow=True,
                          labels=None,
                          explode = explode_list,
                          pctdistance=1.1)
plt.axis('equal')
plt.legend(labels=df_continent.index,loc='upper right')
plt.title('Imigrants to canada by continet from 1980 - 2013')
plt.show()
```

### Asian countries imigrants..

```python
df_asian = df.loc[df.Continent=='Asia']
df_asian=df_asian.reset_index()
df_asian =df_asian.groupby('Region',axis=0).sum()
df_asian.head()
```

```python

colors_list = ['gold', 'yellow', 'lightcoral', 'lightskyblue', 'lightgreen']
explode_list = [0.1,0,0,0,0.2]

df_asian['Total'].plot(kind='pie',figsize=(12,8),
                       autopct='%1.1f%%',
                       startangle=35,
                       pctdistance=1.3,
                       labels=None,explode=explode_list)
plt.legend(df_asian.index,loc='upper left')
plt.axis('equal')
plt.title('Imigration to canada from Asian Countries from 1980 - 2013',y=1)
plt.show()
```

### Scatter Plot.

```python
df_tot = pd.DataFrame(df[years].sum(axis=0))
df_tot.index = map(int,df_tot.index)
df_tot.reset_index(inplace=True)
df_tot.columns = ['Year','Total_Imigrants']
df_tot.head()
```

```python
df_tot.plot(kind='scatter',x='Year',y='Total_Imigrants',figsize=(8,6))\
    .set(title='Total imigrants to canada over 1980 - 2013')
plt.show()  
```

```python
x = df_tot['Year']
y = df_tot['Total_Population']

fit = np.polyfit(x,y,deg=1)
fit
```

Drawing a regression line

```python
df_tot.plot(kind='scatter',x='Year',y='Total_Population',figsize=(10,6),color='darkblue')\
    .set(title='Total imigrants to canada over 1980 - 2013')
plt.plot(x,fit[0]*x+fit[1],color='r')
plt.annotate('y={0:.0f} * X + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))
plt.show()  
```

```python
df_tot_2 = pd.DataFrame(df.loc[['Denmark', 'Norway','Sweden'],years].sum(axis=0))
df_tot_2.index = map(int,df_tot_2.index)
df_tot_2.reset_index(inplace=True)
df_tot_2.columns = ['Year','No of Imigrants']
df_tot_2.plot(kind='scatter',x='Year',y='No of Imigrants',
              figsize=(10,6)).set(title='Imigrants to Canada over 1980 - 2013 from [Denmark,Norway and Sweeden]')
x = df_tot_2['Year']
y = df_tot_2['No of Imigrants']
#Plotting the regression line..
fit = np.polyfit(x,y,deg=1)


plt.plot(x,fit[0]*x+fit[1])
plt.show()
```

```python
df2 = df[years].transpose()

df2.index = map(int,df2.index)
df2.index.name = 'Year'
df2.reset_index(inplace=True)
df2.head()
```

```python
df_ind = (df2['India'] - df2['India'].min() )/(df2['India'].max()-df2['India'].min())
df_china = (df2['China'] - df2['China'].min() )/(df2['China'].max()-df2['China'].min())
```

```python
#India..

ax0 = df2.plot(kind='scatter',x='Year',y='India',
               figsize=(12,8),color='green',
               alpha=0.6,s=df_ind*300*5,
              xlim=(1975,2015))
#China
ax1 = df2.plot(kind='scatter',x='Year',y='China',
               figsize=(12,8),color='red',
               alpha=0.3,s=df_china*500*5,
              xlim=(1975,2015),ax=ax0)

plt.title('Imigrants to canada from china and india form 1980 - 2013')
plt.ylabel('Year')
plt.ylabel('No of imigrants')
plt.legend(['India','China'],loc='upper left')
```
