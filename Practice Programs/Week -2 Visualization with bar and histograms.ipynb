```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Canada.csv',index_col=0)
```

```python
df.rename({'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True,axis=1)
```

```python
df.drop(columns=['Type', 'Coverage', 'AREA', 'REG','DEV'],axis=1,inplace=True)
df.head()
```

```python
all(isinstance(column,str) for column in df.columns)

#lets set the index column as the index..

df.index = df.Country
```

```python
df.shape
```

```python
years = list(map(str,range(1980,2014)))
print(years)
```

```python
### sort values by the top imigrants from all countries
```

```python
df['Total'] = df[years].sum(axis=1)
df.sort_values(by='Total',ascending=False,inplace=True)
top_imigrants = df[years].head()
df.head()
```

```python
top_imigrants = top_imigrants.transpose()
top_imigrants.plot(kind='area',alpha=0.30,stacked=False,figsize=(20,10))
plt.title('Top 5 imigrants to Canda from 1980 - 2013')
plt.xlabel('Years')
plt.ylabel('No. of imigrants')
```

```python
#least imigrants ..
least_impgrants  = df[years].tail()
least_impgrants = least_impgrants.transpose()

ax = least_impgrants.plot(kind='area',alpha=0.5,stacked=False,figsize=(20,10))
ax.set_title('Least imigrants to candada from 1980-2013')
ax.set_xlabel('years')
ax.set_ylabel('No of imigrants')
```

```python
t_df = df.loc[['India','China','South Africa','United States of America'],years]
t_df=t_df.transpose()
```

```python
t_df.plot(kind='hist', figsize=(10, 6))

plt.title('Histogram of Immigration from India ,China, Australia and South Africa from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

```python
count,bins = np.histogram(t_df,bins=20)

t_df.plot(kind='hist', figsize=(20, 10),bins=20,alpha=0.6,xticks=bins,yticks=count,color=['coral', 'darkslateblue', 'mediumseagreen'])

plt.title('Histogram of Immigration from India ,China, Australia and South Africa from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

```python
count,edges = np.histogram(t_df,20)
xmin = edges[0]-100
xmax = edges[-1]+100

t_df.plot(kind='hist', figsize=(20, 10),bins=20,alpha=0.6,xticks=bins,yticks=count,
          color=['coral', 'darkslateblue', 'mediumseagreen','lightblue'],xlim=(xmin,xmax))

plt.title('Histogram of Immigration from India ,China, Australia and South Africa from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

### Barplot

```python
df_india = df.loc[['China','India'],years].transpose()

df_india.plot(kind='bar',alpha=0.6,figsize=(20,8),color=['red','darkblue'])
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('China and India Immigrants to Canada from 1980 to 2013')
```

```python
df_iceland = df.loc['Iceland',years]
df_iceland.plot(kind='bar',alpha=0.6,figsize=(20,8),rot=90,color='r')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('China and India Immigrants to Canada from 1980 to 2013')

plt.annotate('',
             xy=(32,70),
             xytext=(28,20),
             xycoords='data',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='b', lw=2)
            )
```

```python
### Top 15 countries  imigrants to canada.
c = df.head(15).index
top_15 = df.loc[c,'Total'].sort_values(ascending=True)
top_15.plot(kind='barh',figsize=(12,12),alpha=0.6,stacked=False,color=['mediumseagreen'])
plt.title('Top 15 countries imigrants to canada')
plt.xlabel('No of imigrants')
plt.ylabel('Country')

for i,value in enumerate(top_15):
    label = format(int(value), ',')
    plt.annotate(label, xy=(value - 47000, i - 0.1), color='black')
plt.show()
```
