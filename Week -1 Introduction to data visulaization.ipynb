### Plotting in pandas...

```python
haiti = df_canada.loc['Haiti',years]
haiti.plot()
```

```python
haiti.index =  haiti.index.map(int)
haiti.plot(kind='line',color='b')
plt.title("Imigration from Haiti")
plt.xlabel('Years')
plt.ylabel('No of imigrents')


#lets add the name at 2010

plt.text(2000,6000,'2010 Earth Quake')
plt.show()
```

```python
df_CI = df_canada.loc[['Afghanistan','India','China','United States of America'],years].transpose()
df_CI.index = df_CI.index.map(int)
df_CI.plot(kind='line')
plt.title('INdia vs China Imigrants...')
plt.xlabel('Years')
plt.ylabel('No of imigrants')
plt.text(2002,25000,'2010 earth quake')
```

```python
tops = df_canada.loc[:,'Total'].sort_values(ascending=False).head(5).index
top_im = df_canada.loc[tops,years].transpose()
top_im.index = top_im.index.map(int)
top_im.plot(kind='line')
plt.title('Top imigrants to Candada in 2010')
plt.xlabel('Years')
plt.ylabel('No of imigrants..')
plt.figure(figsize=(20,10))
plt.show()
```
